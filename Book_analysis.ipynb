# Most used words in book

```python
pattern = re.compile("[a-zA-Z]+")
finding = re.findall(pattern,book.lower())
finding
```

```python
d={}
for word in finding:
    if word in d.keys():
        d[word]=d[word]+1
    else:
        d[word]=1
```

```python
d_list=[(value, key) for (key,value) in d.items()]
sorted(d_list,reverse = True)
```

```python
pattern = re.compile("*'+[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.+'*")
finding = re.findall(pattern,book)
finding
```

```python
def find(word):
    if word in book:
        return book.count(word)
    else:
        print(f'The book does not contain the word "{word}"')
```

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern,book.lower())
findings[:5]
```

```python
d={}
for word in findings:
    if word in d.keys() and len(word)<6:
        d[word] = d[word] + 1
    else:
        d[word] = 1
        
```

```python
d_list = [(value,key) for (key,value) in d.items()]
d_list = sorted(d_list,reverse=True)
d_list[:5]
```

```python
import nltk

from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

```python
filtered_words=[]
for count ,word in d_list:
    if word not in english_stopwords:
        filtered_words.append((word,count))
```

# Sentiment analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
score = analyzer.polarity_scores("Hey you are looking beautiful today , nice shoes!!")
```

```python
print(score)
```

```python
analyzer.polarity_scores(book)
```

# Chapter sentiment analysis

```python
import re 
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern,book)
```

```python
chapters = chapters[1:]
```

```python
for no , chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(f"CHP - {no+1} {scores}")    
```
